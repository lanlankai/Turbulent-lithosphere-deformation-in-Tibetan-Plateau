# Turbulent lithosphere deformation in the Tibetan Plateau
We present the GPS data and the corresponding structure-function analysis.

<div align=center> <img src="./Fig01.png" style="width: 500px;"></div>
Illustration of the GPS velocity in the Tibetan Plateau. Figure taken from Phys. Rev. E. 99, 062122 (2019).

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.colors import LightSource, Normalize

A=np.load('Tibetan.npz')
Lon=A['Lon']
Lat=A['Lat']
u=A['u']
v=A['v']
Lon_etop=A['Lon_etop']
Lat_etop=A['Lat_etop']
Etop=A['Etop']
c1x=A['c1x']
c1y=A['c1y']
c2x=A['c2x']
c2y=A['c2y']
```

```python
#reproduce fig01 in PRE paper.

plt.clf()
ax=plt.gca()
cmap = plt.cm.terrain
ls = LightSource(315, 45)

rgb = ls.shade(Etop, cmap)
cax=ax.imshow(rgb,
       interpolation='bilinear',
	    origin='lower',
		vmin=Etop.min(),
		vmax=Etop.max(),
		cmap=plt.cm.viridis,
		extent=(Lon_etop.min(), Lon_etop.max(),Lat_etop.min(), Lat_etop.max()),
		alpha=0.8,
		)
Q=u*u+v*v;
Q=np.sqrt(Q)
colors = np.random.rand(553)
ax.scatter(Lon, Lat,s=Q*5e1,c=colors,alpha=0.8,lw=1,edgecolors='none',cmap=plt.cm.jet)
u1=u/Q
v1=v/Q
ax.quiver(Lon,Lat,u1,v1,#Q,
 # linewidths=0.5,
 units='x',
 # pivot='tip',
 width=0.1,
 scale=1,
 )


ax.set_xlabel(r'Longitude',fontsize=14)
ax.set_ylabel(r'Latitude',fontsize=14)
ax.set_xticks([70.000000, 75.000000, 80.000000, 85.000000, 90.000000, 95.000000, 100.000000, 105.000000, 110.000000])
ax.set_xticklabels([r'$70 $', r'$75 $', r'$80 $', r'$85 $', r'$90 $', r'$95 $', r'$100$', r'$105$', r'$110$'])
ax.set_yticks([25.000000, 30.000000, 35.000000, 40.000000, 45.000000])
ax.set_yticklabels([r'$25$', r'$30$', r'$35$', r'$40$'])
ax.set_xlim([74.350000, 108.716667])
ax.set_ylim([22.716667, 41.950000])

ann = ax.annotate("",
                  xy=(95, 30), xycoords='data',
                  xytext=(96,30), textcoords='data',
                  size=15, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"),
                  arrowprops=dict(arrowstyle="<|-",
                                  connectionstyle="arc3,rad=0",
								  shrinkA=5, shrinkB=5,
                                  patchA=None,
                                  patchB=None,
								  color=[1,0,0],
								  lw=1.0,
								#   connectionstyle="angle3,angleA=30,angleB=-30",
                                  fc=[1,0,0]),
                             	)

ann = ax.annotate("",
                  xy=(97, 29.2), xycoords='data',
                  xytext=(98,28), textcoords='data',
                  size=15, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"),
                  arrowprops=dict(arrowstyle="<|-",
                                  connectionstyle="arc3,rad=0",
								  shrinkA=5, shrinkB=5,
                                  patchA=None,
                                  patchB=None,
								  color=[1,0,0],
								  lw=1.0,
								#   connectionstyle="angle3,angleA=30,angleB=-30",
                                  fc=[1,0,0]),
                             	)

ann = ax.annotate("",
                  xy=(92.125, 28), xycoords='data',
                  xytext=(92.4,29), textcoords='data',
                  size=15, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"),
                  arrowprops=dict(arrowstyle="<|-",
                                  connectionstyle="arc3,rad=0",
								  shrinkA=5, shrinkB=5,
                                  patchA=None,
                                  patchB=None,
								  color=[1,0,0],
								  lw=1.0,
								#   connectionstyle="angle3,angleA=30,angleB=-30",
                                  fc=[1,0,0]),
                             	)
ax.plot(c1x,c1y-1.0,'r-',lw=1.5)
ax.plot(c2x[:800],c2y[:800],'r-',lw=1.5)

ann = ax.annotate("",
                  xy=(102, 39), xycoords='data',
                  xytext=(103,38.5), textcoords='data',
                  size=15, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"),
                  arrowprops=dict(arrowstyle="<|-",
                                  connectionstyle="arc3,rad=0",
								  shrinkA=5, shrinkB=5,
                                  patchA=None,
                                  patchB=None,
								  color=[1,0,0],
								  lw=1.0,
								#   connectionstyle="angle3,angleA=30,angleB=-30",
                                  fc=[1,0,0]),
                             	)

ann = ax.annotate("",
                  xy=(103.8, 37.5), xycoords='data',
                  xytext=(104.25,36.5), textcoords='data',
                  size=15, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"),
                  arrowprops=dict(arrowstyle="<|-",
                                  connectionstyle="arc3,rad=0",
								  shrinkA=5, shrinkB=5,
                                  patchA=None,
                                  patchB=None,
								  color=[1,0,0],
								  lw=1.0,
								#   connectionstyle="angle3,angleA=30,angleB=-30",
                                  fc=[1,0,0]),
                             	)
ann = ax.annotate("",
                  xy=(100, 39.5), xycoords='data',
                  xytext=(101,39.5), textcoords='data',
                  size=15, va="center", ha="center",
                  bbox=dict(boxstyle="round4", fc="w"),
                  arrowprops=dict(arrowstyle="<|-",
                                  connectionstyle="arc3,rad=0",
								  shrinkA=5, shrinkB=5,
                                  patchA=None,
                                  patchB=None,
								  color=[1,0,0],
								  lw=1.0,
								#   connectionstyle="angle3,angleA=30,angleB=-30",
                                  fc=[1,0,0]),
                             	)
ax.plot([78,83],[24,24],'-',color=[1,1,1],lw=4)
ax.text(79,24.5,r'$500$\,km',fontsize=10,color=[1,1,1])

ax.xaxis.set_ticks_position('both')
ax.yaxis.set_ticks_position('both')

plt.savefig('Fig01.pdf',dpi=300,bbox_inches='tight',transparent=True)
plt.savefig('Fig01.png',dpi=300,bbox_inches='tight',transparent=True)
```
